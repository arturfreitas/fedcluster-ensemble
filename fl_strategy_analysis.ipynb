```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
import numpy as np
```

# Import results for all strategies analyzed

```python
def plot_metric_comparison(folder_paths, labels, metric_column, title, csv_filename):
    """
    Function to plot a comparison of a specified metric for multiple results.

    Parameters:
    - folder_paths: List of folder paths containing CSV files with metrics.
    - labels: List of labels corresponding to each folder path.
    - metric_column: The name of the column containing the metric to be plotted.
    - title: Title for the plot.
    - csv_filename: The name of the CSV file containing the metrics (e.g., 'centralized_metrics.csv').
    """
    if len(folder_paths) != len(labels):
        raise ValueError("The number of folder paths must match the number of labels.")

    plt.figure(figsize=(10, 5))

    for folder_path, label in zip(folder_paths, labels):
        # Import the specified CSV file
        metrics = pd.read_csv(Path(folder_path) / csv_filename)
        
        # Extract the specified metric
        metric_values = metrics[metric_column]

        # Plot the metric values
        plt.plot(metric_values, label=label)
    
    # Customize the plot
    plt.title(title, fontsize=16)
    plt.xlabel('Rounds', fontsize=14)
    plt.ylabel(metric_column.replace('_', ' ').capitalize(), fontsize=14)
    
    # Set x-ticks to be spaced out and rotated to avoid overlap
    num_rounds = len(metric_values)
    max_ticks = 20  # Maximum number of ticks to display
    step = max(1, num_rounds // max_ticks)
    plt.xticks(np.arange(0, num_rounds, step=step), rotation=45)
    
    plt.legend(fontsize=12)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
