## Testing Chat Ollama with local models

```python
from langchain_ollama import ChatOllama
from  langchain_core.messages import HumanMessage, SystemMessage

llm = ChatOllama(model="llama3.2:3b", temperature=0.3)

msgs = [
    SystemMessage(content="Você é um assistente especialista em aprendizado federado."),
    HumanMessage(content="Explique o que é federated learning em uma frase curta.")
]

response = llm.invoke(msgs)
print(response.content)
```
